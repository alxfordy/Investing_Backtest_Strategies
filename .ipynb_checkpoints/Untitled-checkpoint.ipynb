```python
import yfinance as yf
import mplfinance as mpf

btc = yf.Ticker("BTC-USD")
hist = btc.history(period="max", interval="1d")
```

```python
mpf.plot(hist)
```

```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
import datetime
import pandas as pd


class DCAInvesting(Strategy):
    
    def init(self):
        self.last_buy = None
        
    def next(self):
        print(f"Last Buy {self.last_buy}, current iteration date: {self.data.index[-1]}")
        if self.last_buy:
            time_between_buys = self.data.index[-1] - self.last_buy
            print(f"Time Since Last Buy: {time_between_buys}")
            if time_between_buys.days >= 30:
                print(f"Period larger than 30 days, buying...")
                self.last_buy = self.data.index[-1]
                self.buy()
                print(f"Equity Currently = {self.equity})
            else:
                pass
        else:
            self.last_buy = self.data.index[-1]
            self.buy() 

class ValueInvesting(Strategy):
    INCREASE_PER_MONTH = 100
    
    def init(self):
        print(f"Initiating Value Investing Strategy...")
        self.last_buy = None
        self.thirty_day_intervals_past = 1

    def next(self):
        print(f"Last Buy {self.last_buy}, current iteration date: {self.data.index[-1]}")
        if self.last_buy:
            time_between_buys = self.data.index[-1] - self.last_buy
            print(f"Time Since Last Buy: {time_between_buys}")
            if time_between_buys.days >= 30:
                self.thirty_day_intervals_past += 1
                investment_period_value = INCREASE_PER_MONTH * self.thirty_day_intervals_past
                actual_value = self.equity
                print(f"Equity Currently = {actual_value}, should have {investment_period_value}")
                self.last_buy = self.data.index[-1]
                self.buy(size=1)
            else:
                pass
        else:
            self.last_buy = self.data.index[-1]
            print(f"First Pass, Equity Currently = {self.equity}")
            self.buy(size=1) 
```

```python
VA_BTC = Backtest(hist, ValueInvesting,
              cash=100, commission=.002,
              exclusive_orders=False)

output = VA_BTC.run()
```
